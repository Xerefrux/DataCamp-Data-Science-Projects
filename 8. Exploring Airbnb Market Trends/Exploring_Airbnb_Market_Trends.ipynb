```python
# Import necessary packages
import pandas as pd
import numpy as np
```

```python
airbnb_price = pd.read_csv('E:\\Data Science Course\\Projects\\8. Exploring Airbnb Market Trends\\airbnb_price.csv')
airbnb_room_type = pd.read_excel('E:\\Data Science Course\\Projects\\8. Exploring Airbnb Market Trends\\airbnb_room_type.xlsx', engine='openpyxl')
airbnb_last_review = pd.read_csv('E:\\Data Science Course\\Projects\\8. Exploring Airbnb Market Trends\\airbnb_last_review.tsv', sep='\t')
```

```python
print(airbnb_price.head())
print('\n')
print(airbnb_room_type.head())
print('\n')
print(airbnb_room_type.head())
```

```python
listings = pd.merge(airbnb_price, airbnb_room_type, on='listing_id')
listings = pd.merge(listings, airbnb_last_review, on='listing_id')
```

```python
print(listings.head())
```

```python
listings['last_review_date'] = pd.to_datetime(listings['last_review'], format='%B %d %Y')
first_reviewed = listings['last_review_date'].min()
last_reviewed = listings['last_review_date'].max()
```

```python
print(first_reviewed)
print('\n')
print(last_reviewed)
```

```python
listings['room_type'] = listings['room_type'].str.lower()
private_room_count = listings[listings['room_type'] == 'private room'].shape[0]
```

```python
print(private_room_count)
```

```python
listings['price_clean'] = listings['price'].str.replace(' dollars', '').astype(float)
avg_price = listings['price_clean'].mean()
```

```python
print(avg_price)
```

```python
review_dates = pd.DataFrame({
    'first_reviewed': [first_reviewed],
    'last_reviewed': [last_reviewed],
    'nb_private_rooms': [private_room_count],
    'avg_price': [round(avg_price, 2)]
})
```

```python
print(review_dates)
```
