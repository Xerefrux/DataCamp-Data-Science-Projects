## Task 2: Identify and replace missing values, Clean categorical and text data by manipulating strings, Convert values between data types 

Before you fit any models, you will need to make sure the data is clean. 

The table below shows what the data should look like. 

Create a cleaned version of the dataframe. 

 - You should start with the data in the file "house_sales.csv". 

 - Your output should be a dataframe named `clean_data`. 

 - All column names and values should match the table below.


| Column Name | Criteria                                                |
|-------------|---------------------------------------------------------|
| house_id    | Nominal. </br> Unique identifier for houses. </br>Missing values not possible. |
| city        | Nominal. </br>The city in which the house is located. One of 'Silvertown', 'Riverford', 'Teasdale' and 'Poppleton' </br>Replace missing values with "Unknown". |
| sale_price  | Discrete. </br>The sale price of the house in whole dollars. Values can be any positive number greater than or equal to zero.</br>Remove missing entries. |
| sale_date   | Discrete. </br>The date of the last sale of the house. </br>Replace missing values with 2023-01-01. |
| months_listed  | Continuous. </br>The number of months the house was listed on the market prior to its last sale, rounded to one decimal place. </br>Replace missing values with mean number of months listed, to one decimal place. |
| bedrooms    | Discrete. </br>The number of bedrooms in the house. Any positive values greater than or equal to zero. </br>Replace missing values with the mean number of bedrooms, rounded to the nearest integer. |
| house_type   | Ordinal. </br>One of "Terraced", "Semi-detached", or "Detached". </br>Replace missing values with the most common house type. |
| area      | Continuous. </br>The area of the house in square meters, rounded to one decimal place. </br>Replace missing values with the mean, to one decimal place. |

```python
# Use this cell to write your code for Task 2
import pandas as pd

# Load the data
clean_data = pd.read_csv(r"E:\Data Science Course\Projects\15. Practical Exam - House Sales\house_sales.csv")

# Remove rows with missing sale_price (must be done first before calculating means)
clean_data = clean_data.dropna(subset=['sale_price'])

# Replace missing city values with "Unknown" (missing values are '--' strings)
clean_data['city'] = clean_data['city'].replace('--', 'Unknown')

# Replace missing sale_date values with "2023-01-01"
clean_data['sale_date'] = clean_data['sale_date'].fillna("2023-01-01")

# Replace missing months_listed with mean (rounded to 1 decimal place)
months_listed_mean = round(clean_data['months_listed'].mean(), 1)
clean_data['months_listed'] = clean_data['months_listed'].fillna(months_listed_mean)

# Replace missing bedrooms with mean (rounded to nearest integer)
bedrooms_mean = round(clean_data['bedrooms'].mean())
clean_data['bedrooms'] = clean_data['bedrooms'].fillna(bedrooms_mean)

# Clean house_type - standardize abbreviations before handling missing values
clean_data['house_type'] = clean_data['house_type'].replace({
    'Det.': 'Detached',
    'Semi': 'Semi-detached',
    'Terr.': 'Terraced'
})

# Replace missing house_type with the most common house type
most_common_house_type = clean_data['house_type'].mode()[0]
clean_data['house_type'] = clean_data['house_type'].fillna(most_common_house_type)

# Clean the area column - remove " sq.m." suffix and convert to numeric
clean_data['area'] = clean_data['area'].str.replace(' sq.m.', '', regex=False)
clean_data['area'] = pd.to_numeric(clean_data['area'], errors='coerce')

# Replace missing area with mean (rounded to 1 decimal place)
area_mean = round(clean_data['area'].mean(), 1)
clean_data['area'] = clean_data['area'].fillna(area_mean)

clean_data
```

## Task 3: Aggregate numeric, categorical variables and dates by groups 

The team at RealAgents have told you that they have always believed that the number of bedrooms is the biggest driver of house price. 

Producing a table showing the difference in the average sale price by number of bedrooms along with the variance to investigate this question for the team.

 - You should start with the data in the file 'house_sales.csv'.

 - Your output should be a data frame named `price_by_rooms`. 

 - It should include the three columns `bedrooms`, `avg_price`, `var_price`. 

 - Your answers should be rounded to 1 decimal place.   

```python
#Use this cell to write your code for Task 3
import pandas as pd

# Load the data
df = pd.read_csv(r"E:\Data Science Course\Projects\15. Practical Exam - House Sales\house_sales.csv")

# Group by bedrooms and calculate average and variance of sale_price
price_by_rooms = df.groupby('bedrooms')['sale_price'].agg(['mean', 'var']).reset_index()

# Rename columns to match requirements
price_by_rooms.columns = ['bedrooms', 'avg_price', 'var_price']

# Round to 1 decimal place
price_by_rooms['avg_price'] = price_by_rooms['avg_price'].round(1)
price_by_rooms['var_price'] = price_by_rooms['var_price'].round(1)

price_by_rooms
```

## Task 4: Implement standard modeling approaches for supervised learning problems

Fit a baseline model to predict the sale price of a house.

 1. Fit your model using the data contained in “train.csv” </br></br>

 2. Use “validation.csv” to predict new values based on your model. You must return a dataframe named `base_result`, that includes `house_id` and `price`. The price column must be your predicted values.

```python
# Use this cell to write your code for Task 4
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load the training and validation data
train = pd.read_csv(r"E:\Data Science Course\Projects\15. Practical Exam - House Sales\train.csv")
validation = pd.read_csv(r"E:\Data Science Course\Projects\15. Practical Exam - House Sales\validation.csv")

# Select features for baseline model (numeric features)
features = ['months_listed', 'bedrooms', 'area']

# Prepare training data
X_train = train[features]
y_train = train['sale_price']

# Prepare validation data
X_val = validation[features]

# Fit baseline model (Linear Regression)
model = LinearRegression()
model.fit(X_train, y_train)

# Predict on validation set
predictions = model.predict(X_val)

# Create result dataframe with house_id and price
base_result = pd.DataFrame({
    'house_id': validation['house_id'],
    'price': predictions
})

base_result
```

## Task 5: Implement standard modeling approaches for supervised learning problems

Fit a comparison model to predict the sale price of a house.

 1. Fit your model using the data contained in “train.csv” </br></br>

 2. Use “validation.csv” to predict new values based on your model. You must return a dataframe named `compare_result`, that includes `house_id` and `price`. The price column must be your predicted values.

```python
# Use this cell to write your code for Task 5
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder

# Load the training and validation data
train = pd.read_csv(r"E:\Data Science Course\Projects\15. Practical Exam - House Sales\train.csv")
validation = pd.read_csv(r"E:\Data Science Course\Projects\15. Practical Exam - House Sales\validation.csv")

# Create a copy for feature engineering
train_df = train.copy()
val_df = validation.copy()

# Encode categorical features
le_city = LabelEncoder()
le_house_type = LabelEncoder()

# Fit on combined data to ensure all categories are captured
all_cities = pd.concat([train_df['city'], val_df['city']])
all_house_types = pd.concat([train_df['house_type'], val_df['house_type']])

le_city.fit(all_cities)
le_house_type.fit(all_house_types)

train_df['city_encoded'] = le_city.transform(train_df['city'])
train_df['house_type_encoded'] = le_house_type.transform(train_df['house_type'])

val_df['city_encoded'] = le_city.transform(val_df['city'])
val_df['house_type_encoded'] = le_house_type.transform(val_df['house_type'])

# Select features for comparison model
features = ['months_listed', 'bedrooms', 'area', 'city_encoded', 'house_type_encoded']

# Prepare training data
X_train = train_df[features]
y_train = train_df['sale_price']

# Prepare validation data
X_val = val_df[features]

# Fit comparison model (Random Forest)
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predict on validation set
predictions = model.predict(X_val)

# Create result dataframe with house_id and price
compare_result = pd.DataFrame({
    'house_id': validation['house_id'],
    'price': predictions
})

compare_result
```

```python
# Verify RMSE for both models using cross-validation on training data
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder

# Load the training data
train = pd.read_csv(r"E:\Data Science Course\Projects\15. Practical Exam - House Sales\train.csv")

# Task 4 - Baseline Model (Linear Regression)
features_base = ['months_listed', 'bedrooms', 'area']
X_train_base = train[features_base]
y_train = train['sale_price']

base_model = LinearRegression()
base_scores = cross_val_score(base_model, X_train_base, y_train, cv=5, scoring='neg_root_mean_squared_error')
base_rmse = -base_scores.mean()

# Task 5 - Comparison Model (Random Forest with encoded categorical features)
train_df = train.copy()
le_city = LabelEncoder()
le_house_type = LabelEncoder()
train_df['city_encoded'] = le_city.fit_transform(train_df['city'])
train_df['house_type_encoded'] = le_house_type.fit_transform(train_df['house_type'])

features_compare = ['months_listed', 'bedrooms', 'area', 'city_encoded', 'house_type_encoded']
X_train_compare = train_df[features_compare]

compare_model = RandomForestRegressor(n_estimators=100, random_state=42)
compare_scores = cross_val_score(compare_model, X_train_compare, y_train, cv=5, scoring='neg_root_mean_squared_error')
compare_rmse = -compare_scores.mean()

print(f"Task 4 - Baseline Model (Linear Regression) RMSE: {base_rmse:,.2f}")
print(f"Task 5 - Comparison Model (Random Forest) RMSE: {compare_rmse:,.2f}")
print()
if base_rmse < 30000 or compare_rmse < 30000:
    print(" At least one model has RMSE below 30,000")
else:
    print("Neither model has RMSE below 30,000")
```
